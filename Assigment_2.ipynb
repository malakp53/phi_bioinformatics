```python
#Part 3: Reshaping Data
cellsize=pd.read_csv('ltee_cell_size.tsv', sep='\t')
print(cellsize.head())
#Check col names
cellsize.info()
```

```python
#Convert the column names using this command
cellsize.columns = (['Generation'] + ['Ara-{}'.format(n) for n in
range(1,7)] + ['Ara+{}'.format(n) for n in range(1,7)])
```

```python
#Convert your cellsize dataframe from wide to long format and rename the
#columns with these commands:
cellsize = pd.melt(cellsize, id_vars=cellsize.columns[0],
value_vars=cellsize.columns[1:14].tolist())
cellsize.columns=["Generation", "Population", "Cell Size"]
```

```python
cellsize
```

```python
# Merge the cellsize data frame with the sampleruns data frame on the
#‘Generation’ and ‘Population’ variables to a variable called ltee
ltee=pd.merge(cellsize,sampleruns, on=["Population", "Generation"], how='right')

ltee.head()
```
