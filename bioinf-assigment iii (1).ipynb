```python
#Import important packages
import GEOparse as gpar
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gse=gpar.get_GEO("GSE22552")

```

```python
#Create value tables
data=pd.concat([gse.gsms[gsm].table.set_index("ID_REF").rename(columns={'VALUE':gsm}) for gsm in gse.gsms], axis=1)

#Drop the unsorted samples
data.drop(['GSM559701','GSM559702','GSM559703','GSM559704'], inplace=True, axis=1)

#Drop genes with values < 6
data=data[(data<6).sum(axis=1)<=9]

```

```python
#3. Task 3 – second round of preprocessing (10%)

#a. Summarise replicates (columns that belong to the same condition) using the median

stages={'GSM559689': 'stage 1', 'GSM559690': 'stage 1', 'GSM559691': 'stage 1',
      'GSM559692': 'stage 2','GSM559693': 'stage 2','GSM559694': 'stage 2',
       'GSM559695': 'stage 3','GSM559696': 'stage 3', 'GSM559697': 'stage 3',
      'GSM559698': 'stage 4','GSM559699': 'stage 4', 'GSM559700': 'stage 3' }

df_sum = data_quantile_normalised.groupby(stages, axis=1).median()

df_sum


#b. Calculate z-scores for rows
import scipy.stats as stats
zscores=stats.zscore(df_sum.iloc[0:-1, ])

zscores
#. Plot the first 1000 genes data after z-score transformation. You can use this to do so:
#data.iloc[:1000].T.plot

data_z = df_sum.subtract(df_sum.mean(axis=1), axis=0)
data_z = data_z.divide(df_sum.std(axis=1), axis=0)
data_z.head()
data_z.iloc[:1000].T.plot()
plt.legend([])
```

```python
#4. Task 4 – k-means clustering (20%)
#Perform k-means clustering to group genes into 5 clusters.
from sklearn.cluster import KMeans
%matplotlib inline

model = KMeans(n_clusters=5)
clusters = model.fit_predict(data_z)

data_z[clusters == 0]
np.unique(clusters)
```

```python
#5. Task 5 – visualise clusters (20%)

f, ax = plt.subplots(2,3,figsize=(13,9))

for c in np.unique(clusters):
    data_z[clusters == c].T.plot(ax=ax[int(c/3), c % 3])
    ax[int(c/3), c % 3].legend([])
    ax[int(c/3), c % 3].set_title(f"C{c} ({np.sum(clusters == c)})")

plt.tight_layout()
```
