```python
#Import GEOparse
import GEOparse as gpar
```

```python
#Task 1: load the dataset with the ID GSE22552.
gse=gpar.get_GEO("GSE22552")
```

```python
#Task2 Display (print) the sample IDs (GSMs) included in GSE22552, and the platform ID (GPL)

print(gse.gsms)
type(gse.gsms)
print(gse.gpls)
type(gse.gpls)
```

```python
#Task 3
#Concatenate the “VALUE” columns of the tables of all 16 samples of the dataset to form one single
#pandas dataframe. Make the ID_REF column as the index column of the generated dataframe. 
#Import pandas
import pandas as pd
#Explore table structure
print(gse.gsms["GSM559689"].table)

#Concat> my own code

frames=[] #I created this list to append
for gsm in gse.gsms: #for loop to get each sample table, and append it to the list "frames"
    frames.append(gse.gsms[gsm].table.set_index('ID_REF'))
    
gse_con=pd.concat(frames, axis=1)
gse_con.columns=gse.gsms.keys()  #rename the columns


```

```python
#Explore gse_con
gse_con
```

```python
gse.gpls["GPL570"].table
```

```python
gpls=gse.gpls["GPL570"].table
type(gpls)
gpls.set_index('ID')
```

```python
#Task 4  using pandas merge to merge the two columns “Gene Symbol” and “ENTREZ_GENE_ID” into the data DataFrame

gpls_id= gpls[['ENTREZ_GENE_ID', 'Gene Symbol']].set_index(gpls["ID"])
gpls_id
```

```python
#Task 4  using pandas merge to merge the two columns “Gene Symbol” and “ENTREZ_GENE_ID” into the data DataFrame
data=pd.merge(gse_con,gpls_id,left_index=True, right_index=True, how='left')
```

```python
data
```

```python
#Task 5 , a) Can you keep the first token only while removing the “ /// ” #separator and everything after it? 


#Solution: 1- define a function to remove the '///' separator and everything after
def separate(string):
    return string.split('///')[0]
    
    
#2- use apply function on Gene symbol and Entrez IDs
data['Gene Symbol']=data['Gene Symbol'].astype(str).apply(separate) 
data['ENTREZ_GENE_ID']=data['ENTREZ_GENE_ID'].astype(str).apply(separate) 
```
